# 05 — Lot-size rounding for F&O

NIFTY options trade in lots of 50; BANKNIFTY in lots of 15. A strategy computing a raw share count must round to the nearest lot before placing an order. `round_to_lot` uses banker's rounding; `floor_to_lot` is strict for hard cash budgets.

```python
from ml4t.india.backtest import round_to_lot, floor_to_lot

# Banker's rounding: 75 / 50 = 1.5 -> 2 lots
print('round:', round_to_lot(75, 50))       # 100
print('round:', round_to_lot(76, 50))       # 100 (above half)
print('round:', round_to_lot(74, 50))       # 50  (below half)

# Floor: never exceed budget
print('floor:', floor_to_lot(149, 50))      # 100
```

`round_to_lot` raises if a non-zero request rounds to zero -- so the strategy never silently produces a no-op order:

```python
try:
    round_to_lot(10, 50)
except ValueError as e:
    print('raised:', e)
```
