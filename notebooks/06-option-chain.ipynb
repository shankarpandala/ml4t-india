# 06 — Option chain for NFO

Build an `OptionChain` from instruments dump rows, then filter by moneyness around ATM.

```python
import datetime as dt
from ml4t.india.kite.instruments import InstrumentMeta
from ml4t.india.options import OptionChain

expiry = dt.date(2026, 4, 24)
instruments = []
for strike in [24800, 24900, 25000, 25100, 25200]:
    for typ in ('CE', 'PE'):
        instruments.append(InstrumentMeta(
            instrument_token=int(strike*10),
            exchange_token=0,
            tradingsymbol=f'NIFTY26APR{strike}{typ}',
            name='NIFTY', last_price=0.0, expiry=expiry,
            strike=float(strike), tick_size=0.05, lot_size=50,
            instrument_type=typ, segment='NFO-OPT', exchange='NFO',
        ))

chain = OptionChain.from_instruments(instruments, 'NIFTY', expiry)
print('ATM @ spot 25030:', chain.atm_strike(25030))

calls, puts = chain.around_atm(25030, count=1)
for c in calls: print('CE', c.strike, c.tradingsymbol)
for p in puts: print('PE', p.strike, p.tradingsymbol)
```
