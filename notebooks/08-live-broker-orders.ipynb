# 08 — Live broker orders via KiteBroker

`KiteBroker` extends `IndianBrokerBase` and wraps `AsyncKiteClient`. This notebook drives it with a `FakeKiteClient` so you can run it without a live Kite account.

```python
import asyncio
from ml4t.india.kite.client import KiteClient, AsyncKiteClient
from ml4t.india.kite.fake import FakeKiteClient
from ml4t.india.kite.rate_limit import KiteRateLimiter
from ml4t.india.live import KiteBroker
from ml4t.backtest.types import OrderType

sdk = FakeKiteClient()
sdk.set_margins({'equity': {'available': {'cash': 1_000_000}}})
sync = KiteClient(sdk=sdk, rate_limiter=KiteRateLimiter())
broker = KiteBroker(AsyncKiteClient(sync))

async def demo():
    await broker.connect()
    cash = await broker.get_cash_async()
    print(f'cash: Rs {cash:,.2f}')
    order = await broker.submit_order_async(
        asset='NSE:RELIANCE', quantity=10, order_type=OrderType.MARKET,
    )
    print('order:', order.order_id, order.side, order.quantity)

asyncio.run(demo())
```
