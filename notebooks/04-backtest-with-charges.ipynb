# 04 — Backtest with Indian charges

Two complementary tools:
1. `nse_india_config()` preset -- a `BacktestConfig` with sensible blended-cost defaults.
2. `ZerodhaChargesModel` -- exact Zerodha charges (STT sell-side, GST, state stamp duty, exchange + SEBI fees) for post-hoc fill analysis.

```python
from ml4t.india.backtest import nse_india_config, ZerodhaChargesModel
from ml4t.india.backtest.charges import Segment

# Preset: one-line config for an intraday equity backtest
cfg = nse_india_config(initial_cash=10_000_000.0)
print('commission_rate:', cfg.commission_rate, 'slippage_rate:', cfg.slippage_rate)

# Exact charges example -- 10 shares of RELIANCE @ 2500 on the sell side
model = ZerodhaChargesModel(default_segment=Segment.EQUITY_INTRADAY)
charges = model.calculate('NSE:RELIANCE', -10, 2500.0)
print(f'Charges on 10 x 2500 MIS sell: Rs {charges:.2f}')
```
