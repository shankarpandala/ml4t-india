# 03 — Historical OHLCV via KiteProvider

`KiteProvider` extends upstream `BaseProvider` and fetches candles via `kite.historical_data`. Long date ranges are chunked per Kite's per-interval daily ceiling (minute: 60 days, day: 2000 days, etc.).

This notebook uses a `FakeKiteClient` that returns canned bars.

```python
import datetime as dt
from ml4t.india.kite.fake import FakeKiteClient
from ml4t.india.kite.client import KiteClient
from ml4t.india.kite.instruments import InstrumentsCache
from ml4t.india.data.kite import KiteProvider

sdk = FakeKiteClient()
sdk.set_historical_data('738561', [
    {'date': dt.datetime(2024, 1, 2, 9, 15), 'open': 2500, 'high': 2520,
     'low': 2495, 'close': 2510, 'volume': 100000},
    {'date': dt.datetime(2024, 1, 3, 9, 15), 'open': 2510, 'high': 2530,
     'low': 2505, 'close': 2525, 'volume': 120000},
])

client = KiteClient(sdk=sdk)
# (InstrumentsCache omitted for brevity -- KiteProvider takes one.)
```
