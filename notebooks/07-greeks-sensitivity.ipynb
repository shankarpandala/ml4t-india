# 07 — Black-Scholes Greeks

`compute_greeks` returns delta / gamma / vega / theta / rho for a European option. Uses `py_vollib` if installed; falls back to a `math.erf`-based closed-form.

```python
from ml4t.india.options import compute_greeks

g = compute_greeks(
    flag='CE', spot=25000, strike=25000,
    time_to_expiry=30/365, risk_free_rate=0.07, volatility=0.15,
    prefer_pyvollib=False,
)
print(f'delta={g.delta:.4f}  gamma={g.gamma:.6f}')
print(f'vega={g.vega:.2f}   theta/yr={g.theta:.2f}   rho={g.rho:.2f}')
```

Delta sensitivity to spot moves (useful for ATM delta hedging):

```python
for spot in [24500, 24800, 25000, 25200, 25500]:
    g = compute_greeks('CE', spot, 25000, 30/365, 0.07, 0.15, prefer_pyvollib=False)
    print(f'spot={spot}: delta={g.delta:+.3f}')
```
