# 09 — Streaming ticks with KiteTickerFeed

`KiteTickerFeed` wraps `kiteconnect.KiteTicker`. Subscribe before or after `start()`; on reconnect the subscription map is replayed automatically.

Uses a `FakeTicker` test double so no websocket is opened.

```python
import asyncio
from typing import Any
from ml4t.india.live import KiteTickerFeed

class FakeTicker:
    on_ticks = on_connect = on_close = on_error = None
    def __init__(self, *_): self.connected = False
    def connect(self, threaded=True): self.connected = True
    def close(self, *_, **__): pass
    def subscribe(self, tokens): print('subscribed:', tokens)
    def unsubscribe(self, tokens): pass
    def set_mode(self, mode, tokens): print('mode:', mode, tokens)
    def is_connected(self): return self.connected

feed = KiteTickerFeed('k', 'a', ticker_factory=lambda *_: FakeTicker())
feed.on_ticks(lambda ticks: print('tick batch:', len(ticks)))

async def demo():
    await feed.start()
    feed.subscribe([738561, 2953217], mode='quote')

asyncio.run(demo())
```
